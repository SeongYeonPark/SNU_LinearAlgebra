```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(7)

#cluster size N (3 of these will be used)
N = 10

#cluster width alpha
alpha = 1.1
```

```python
cluster_12 = np.random.rand(N, 2) * alpha - alpha/2 + np.array([1,2])
cluster_31 = np.random.rand(N, 2) * alpha - alpha/2 + np.array([3,1])
cluster_00 = np.random.rand(N, 2) * alpha - alpha/2 + np.array([0,0])

datapoints = np.concatenate([cluster_12, cluster_31, cluster_00], axis = 0)
```

```python
plt.scatter(datapoints[:,0], datapoints[:,1])
```

```python
def norm2(x, y):
    return np.sqrt(np.sum((x-y)**2))
```

```python
z1 = np.array([2,2])
z2 = np.array([1.5,0.1])
z3 = np.array([0.2,0.2])
zs = np.stack((z1, z2, z3), axis = 0)
```

```python
def closest_z_index(x, zs):
    k = np.zeros(3)
    for i in range(3):
        k[i] = norm2(x, zs[i])
    return np.argmin(k)
```

```python
ind = np.zeros(datapoints.shape[0])
```

```python
def show():
    color = ['r', 'g', 'b']
    for j in range(3):
        plt.scatter(datapoints[ind == j][:,0], datapoints[ind == j][:,1], c = color[j])
    plt.scatter(zs[:,0], zs[:,1], s = 50, c = 'orange')
    plt.show()
```

```python
for n in range(20):
    for i in range(datapoints.shape[0]):
        indx = closest_z_index(datapoints[i], zs)
        ind[i] = indx
    for j in range(3):
        zs[j] = np.mean(datapoints[ind == j], axis = 0)
    show()
```
