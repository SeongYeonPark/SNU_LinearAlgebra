```python
import numpy as np
import requests
import matplotlib.pyplot as plt

from PIL import Image, ImageFilter
from io import BytesIO
%matplotlib inline
```

```python
# download image and show
url1 = 'https://imgur.com/sQlc7oZ.jpg'
url2 = 'https://imgur.com/5K2oALo.jpg'
response1 = requests.get(url1)
response2 = requests.get(url2)
img1 = Image.open(BytesIO(response1.content)).convert('L').resize((64, 64))
img2 = Image.open(BytesIO(response2.content)).convert('L').resize((64, 64))
fig, ax = plt.subplots(1,2)
ax[0].imshow(np.asarray(img1), cmap='gray')
ax[1].imshow(np.asarray(img2), cmap='gray')
```

```python
# blur the image
blurry_img1 = img1.filter(ImageFilter.GaussianBlur(1))
blurry_img2 = img2.filter(ImageFilter.GaussianBlur(1))
fig, ax = plt.subplots(1,2)
ax[0].imshow(np.asarray(blurry_img1), cmap='gray')
ax[1].imshow(np.asarray(blurry_img2), cmap='gray')
```

```python
# TODO: implement least-square deblurring algorithm
H, W = 64, 64
blurry_img1=np.array(blurry_img1)
blurry_img2=np.array(blurry_img2)
blurry_img1 = blurry_img1.flatten()
blurry_img2 = blurry_img2.flatten()
Dv = np.zeros(((H-1)*W, H*W))
for i in range((H-1)*W):
    Dv[i,i]=-1
    Dv[i,i+W]=1
Dh = np.zeros(((W-1)*H, W*H))
for k in range(H):
    for j in range(W-1):
        Dh[(W-1)*k+j,W*k+j]=-1
        Dh[(W-1)*k+j,W*k+j+1]=1
I = np.eye(H*W)
#print(I.shape, Dv.shape, Dh.shape)
```

```python
lamb1 = 1.5
t = np.sqrt(lamb1)
A = np.vstack((I, t*Dh, t*Dv))
q, r = np.linalg.qr(A)
r_i = np.linalg.inv(r)
y1 = np.zeros(A.shape[0])
y1[:H*W]=blurry_img1
x1 = np.dot(r_i, np.dot(q.T,y1))
plt.imshow(x1.reshape(H,W), cmap='gray')
```
